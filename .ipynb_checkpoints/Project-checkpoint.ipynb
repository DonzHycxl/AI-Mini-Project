Define Image Scraping and Downloading Functions

```python
# Function to get image URLs from a search URL
def get_image_urls(search_url, max_images=2000):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    urls = []
    page = 0

    while len(urls) < max_images:
        try:
            # Send a GET request to the search URL with pagination
            response = requests.get(search_url + f'&first={page * 50}', headers=headers, timeout=5)
            response.raise_for_status()  # Check for HTTP errors
        except requests.RequestException as e:
            print(f"Request failed: {e}")
            break

        # Parse the HTML content of the response
        soup = BeautifulSoup(response.text, 'html.parser')
        # Find all 'a' tags with class 'iusc' (these contain image data)
        image_elements = soup.find_all('a', class_='iusc')

        if not image_elements:
            break

        for img in image_elements:
            if len(urls) >= max_images:
                break
            m = img.get('m')
            if m:
                try:
                    m = ast.literal_eval(m)
                    if 'murl' in m:
                        urls.append(m['murl'])
                except ValueError:
                    continue

        page += 1
        time.sleep(1)

    return urls

# Function to download images from a list of URLs
def download_images(urls, folder_name):
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)

    for i, url in enumerate(urls):
        try:
            img_data = requests.get(url, timeout=5).content
            image_path = f'{folder_name}/{i + 1}.jpg'
            with open(image_path, 'wb') as handler:
                handler.write(img_data)
        except requests.RequestException as e:
            print(f"Could not download {url}: {e}")
```

Download Images and Split the Dataset

```python
from google.colab import drive
drive.mount('/content/drive')

import os
import shutil
from sklearn.model_selection import train_test_split

# Define your classes and corresponding URLs for image collection
classes = ['German Shepherd', 'Bulldog', 'Labrador Retriever', 'Golden Retriever', 'Siberian Husky']
urls = {
    'German Shepherd': 'https://www.bing.com/images/search?q=German%20Shepherd',
    'Bulldog': 'https://www.bing.com/images/search?q=Bulldog',
    'Labrador Retriever': 'https://www.bing.com/images/search?q=Labrador%20Retriever',
    'Golden Retriever': 'https://www.bing.com/images/search?q=Golden%20Retriever',
    'Siberian Husky': 'https://www.bing.com/images/search?q=Siberian%20Husky'
}

# Path to store datasets
base_folder = '/content/drive/My Drive/Dataset'

# Create directories for each class
for class_name in classes:
    class_folder = os.path.join(base_folder, class_name)
    if not os.path.exists(class_folder):
        os.makedirs(class_folder)

# Fetch and download images for each class
for cat, url in urls.items():
    print(f"Downloading images for {cat}")
    image_urls = get_image_urls(url, max_images=2000)  # Ensure your function supports max_images or similar parameter
    download_images(image_urls, os.path.join(base_folder, cat))

print("Image downloading complete.")

# Function to split dataset into training, validation, and testing
def split_dataset(base_folder, classes, train_ratio=0.7, val_ratio=0.15):
    for class_name in classes:
        class_folder = os.path.join(base_folder, class_name)
        images = os.listdir(class_folder)
        train, test = train_test_split(images, test_size=1-train_ratio)
        val, test = train_test_split(test, test_size=val_ratio/(1-train_ratio))

        for t in train:
            shutil.move(os.path.join(class_folder, t), os.path.join(base_folder, 'train', class_name, t))
        for v in val:
            shutil.move(os.path.join(class_folder, v), os.path.join(base_folder, 'val', class_name, v))
        for tst in test:
            shutil.move(os.path.join(class_folder, tst), os.path.join(base_folder, 'test', class_name, tst))

# Create directories for training, validation, and testing
for split in ['train', 'val', 'test']:
    split_folder = os.path.join(base_folder, split)
    if not os.path.exists(split_folder):
        os.makedirs(split_folder)
    for class_name in classes:
        class_split_folder = os.path.join(split_folder, class_name)
        if not os.path.exists(class_split_folder):
            os.makedirs(class_split_folder)

# Split the dataset
split_dataset(base_folder, classes)
```

```python
import os
from PIL import Image

def delete_corrupted_images(directory):
    for subdir, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(subdir, file)
            try:
                with Image.open(file_path) as img:
                    img.verify()  # Verify that it is, in fact, an image
            except (IOError, SyntaxError):
                print('Deleting corrupted image:', file_path)
                os.remove(file_path)  # Delete the corrupted file

# Call this function with the path to your dataset
delete_corrupted_images('path_to_your_dataset')
```

TRAIN


```python
import tensorflow as tf
from tensorflow.keras.utils import Sequence
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
import os

class CustomDataGenerator(Sequence):
    def __init__(self, directory, batch_size=32, target_size=(224, 224), class_mode='categorical', shuffle=True):
        self.directory = directory
        self.batch_size = batch_size
        self.target_size = target_size
        self.class_mode = class_mode
        self.shuffle = shuffle
        self.classes = os.listdir(directory)
        self.class_indices = {c: i for i, c in enumerate(self.classes)}
        self.data = self._load_filenames()
        if self.shuffle:
            np.random.shuffle(self.data)

    def __len__(self):
        return len(self.data) // self.batch_size

    def __getitem__(self, idx):
        batch_data = self.data[idx * self.batch_size:(idx + 1) * self.batch_size]
        return self._data_generation(batch_data)

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.data)

    def _load_filenames(self):
        data = []
        for class_name in self.classes:
            class_dir = os.path.join(self.directory, class_name)
            filenames = [os.path.join(class_dir, f) for f in os.listdir(class_dir) if f.endswith(('.png', '.jpg', '.jpeg'))]
            for fname in filenames:
                data.append((fname, self.class_indices[class_name]))
        return data

    def _data_generation(self, batch_data):
        images = []
        labels = []
        for path, label in batch_data:
            try:
                img = load_img(path, target_size=self.target_size)
                img = img_to_array(img)
                img /= 255.0
                images.append(img)
                labels.append(label)
            except Exception as e:
                print(f"Error loading image {path}: {e}")
                continue
        X = np.array(images)
        y = np.array(labels)
        if self.class_mode == 'categorical':
            y = tf.keras.utils.to_categorical(y, num_classes=len(self.classes))
        return X, y

# Instantiate your custom generators
train_generator = CustomDataGenerator(train_dir, batch_size=32, target_size=(224, 224), class_mode='categorical')
val_generator = CustomDataGenerator(val_dir, batch_size=32, target_size=(224, 224), class_mode='categorical')
test_generator = CustomDataGenerator(test_dir, batch_size=32, target_size=(224, 224), class_mode='categorical')

# Now, use these in your model training and evaluation blocks
epochs = 50
history_resnet50 = resnet50_model.fit(train_generator, epochs=epochs, validation_data=val_generator)
history_densenet121 = densenet121_model.fit(train_generator, epochs=epochs, validation_data=val_generator)
history_mobilenetv3 = mobilenetv3_model.fit(train_generator, epochs=epochs, validation_data=val_generator)

# Evaluate models
print("Evaluating ResNet50")
resnet50_model.evaluate(test_generator)
print("Evaluating DenseNet121")
densenet121_model.evaluate(test_generator)
print("Evaluating MobileNetV3")
mobilenetv3_model.evaluate(test_generator)
```

MAP

```python
import numpy as np
from sklearn.metrics import average_precision_score

def calculate_mAP(model, data_generator):
    y_true = []
    y_pred = []
    for i in range(len(data_generator)):
        X, y = data_generator[i]
        predictions = model.predict(X)
        y_true.extend(y)
        y_pred.extend(predictions)

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    # Calculate average precision for each class
    ap_scores = []
    for i in range(y_true.shape[1]):
        ap = average_precision_score(y_true[:, i], y_pred[:, i])
        ap_scores.append(ap)
        print(f"Average Precision for class {i}: {ap:.4f}")

    # Calculate mean Average Precision
    mAP = np.mean(ap_scores)
    return mAP

# Calculate mAP for each model
print("Calculating mAP for ResNet50")
resnet50_mAP = calculate_mAP(resnet50_model, test_generator)
print(f"ResNet50 mAP: {resnet50_mAP:.4f}")

print("Calculating mAP for DenseNet121")
densenet121_mAP = calculate_mAP(densenet121_model, test_generator)
print(f"DenseNet121 mAP: {densenet121_mAP:.4f}")

print("Calculating mAP for MobileNetV3")
mobilenetv3_mAP = calculate_mAP(mobilenetv3_model, test_generator)
print(f"MobileNetV3 mAP: {mobilenetv3_mAP:.4f}")
```

 Visualize Training Results

```python
import matplotlib.pyplot as plt

def plot_training_history(history, model_name):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(len(acc))

    plt.figure(figsize=(14, 7))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.title(f'{model_name} Training and Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.title(f'{model_name} Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

# Plot history for each model
plot_training_history(history_resnet50, 'ResNet50')
plot_training_history(history_densenet121, 'DenseNet121')
plot_training_history(history_mobilenetv3, 'MobileNetV3')
```

 Evaluate Models with Confusion Matrix and Classification Report

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

def evaluate_model_with_confusion_matrix_and_report(model, data_generator, model_name):
    y_true = []
    y_pred = []
    for i in range(len(data_generator)):
        X, y = data_generator[i]
        predictions = model.predict(X)
        y_true.extend(np.argmax(y, axis=1))
        y_pred.extend(np.argmax(predictions, axis=1))

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    # Generate confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=data_generator.classes, yticklabels=data_generator.classes)
    plt.title(f'{model_name} Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

    # Generate classification report
    print(f'{model_name} Classification Report:')
    print(classification_report(y_true, y_pred, target_names=data_generator.classes))

# Evaluate and plot confusion matrix and classification report for each model
print("Evaluating ResNet50")
evaluate_model_with_confusion_matrix_and_report(resnet50_model, test_generator, 'ResNet50')

print("Evaluating DenseNet121")
evaluate_model_with_confusion_matrix_and_report(densenet121_model, test_generator, 'DenseNet121')

print("Evaluating MobileNetV3")
evaluate_model_with_confusion_matrix_and_report(mobilenetv3_model, test_generator, 'MobileNetV3')
```
